```python
from scipy.stats import norm

def z_critical(alpha, test_type):
    """
    alpha: significance level (e.g. 0.05)
    test_type: 'left', 'right', or 'two'
    """

    if test_type == 'left':
        return norm.ppf(alpha)

    elif test_type == 'right':
        return norm.ppf(1 - alpha)

    elif test_type == 'two':
        z = norm.ppf(1 - alpha/2)
        return -z, z

    else:
        raise ValueError("test_type must be 'left', 'right', or 'two'")
```

```python
alpha = 0.05
zc = z_critical(alpha, 'left')
print("Z critical value:", zc)
```

```python
print("Z critical value:", z_critical(0.05, 'right'))
```

```python
print("Z critical values:", z_critical(0.05, 'two'))
```

```python
def z_test_decision(z_calc, alpha, test_type):
    zc = z_critical(alpha, test_type)

    if test_type == 'left':
        return "Reject H0" if z_calc <= zc else "Fail to Reject H0"

    elif test_type == 'right':
        return "Reject H0" if z_calc >= zc else "Fail to Reject H0"

    elif test_type == 'two':
        return "Reject H0" if z_calc <= zc[0] or z_calc >= zc[1] else "Fail to Reject H0"
```

```python
print(z_test_decision(-1.5, 0.05, 'left'))
```
