A coaching institute claims that students score an average of 75 marks in a competitive exam. A sample of 10 students has a mean score of 72 with a sample standard deviation of 6.

At 5% significance level, test the claim using the critical value method.

```python
import math
from scipy import stats

# Given values
mu = 75
xbar = 72
s = 6
n = 10
alpha = 0.05

# Degrees of freedom
df = n - 1

# Critical t value for two-tailed test
tc = stats.t.ppf(1 - alpha/2, df)

# Standard Error
SE = s / math.sqrt(n)

# Critical interval
lower_CV = mu - tc * SE
upper_CV = mu + tc * SE

print("Critical Value Method Output")
print("Lower Bound =", round(lower_CV, 2))
print("Upper Bound =", round(upper_CV, 2))

if lower_CV <= xbar <= upper_CV:
    print("\nSample mean lies inside the critical range.")
    print("Conclusion = Fail to reject H0 (Institute claim is not rejected)")
else:
    print("\nSample mean lies outside the critical range.")
    print("Conclusion = Reject H0 (Institute claim is rejected)")
```
